# **6. Data Pre-Processing**
---

- This section is emphasised on performing data manipulation over unstructured data for further processing and analysis.

- To modify unstructured data to strucuted data you need to verify and manipulate the integrity of the data by:
  - Handling missing data,

  - Handling redundant data,

  - Handling inconsistent data,

  - Handling outliers,

  - Handling typos

```python

# HTAG is a continuous feature, we can the replace null values with rounding off the mean of the feature.
value=data['HTAG'].mean()
value=np.round(value)
data['HTAG'] = data['HTAG'].fillna(value = value)
#data['HTAG'] = data['HTAG'].fillna(value = np.round(data['HTAG'].mean())

# HTHG is a continuous feature, we can the replace null values with rounding off the mean of the feature.
value = data['HTHG'].mean()
value = np.round(value)
data['HTHG'] = data['HTHG'].fillna(value = value)

# HTR is a categorical feature, we can the replace null values with mode of the feature.
data['HTR'] = data['HTR'].fillna(value=data['HTR'].mode()[0])
 
#changing data type of date column
data["Date"]= pd.to_datetime(data["Date"])

#dropping Div axis as it has a constant value that is highly insignificant
data.drop('Div',axis=1,inplace=True)
```

```python
#Function to plot pie chart
def plotPieChart(data=None, column=None):
    # Create a figure size of 15 X 7 inches.
    figure = plt.figure(figsize=[15, 7])

    # Seeting searbon style in pyplot interface
    plt.style.use('seaborn')
    sns.set_style("whitegrid")

    # Extract labels and values using .value_counts() function.
    extract = data[column].value_counts().sort_index()
    labels, sizes = extract.index, extract.values

    # Initiate an explode list of categories
    explode_list = np.ones(len(data[column].unique()))/10

    # Using .pie() function to pass the xaxis and yaxis data
    plt.pie(x=sizes, labels=labels, explode=explode_list, autopct='%1.1f%%', 
            startangle=180, shadow=True, wedgeprops=dict(width=0.15))

    plt.axis('equal')

    # Set title label of size 16.
    plt.title(label='Pie Distribution of '+column, size=16)

    # Display the output
    plt.show()

 
```

```python
#Function to display performance of user entered team
def teamperformance(team=None):
    #filtering data
    data1=data[data['HomeTeam']== team]
    data2=data[data['AwayTeam']== team]

    # Instantiate figure and axes object
    f, ax = plt.subplots(nrows=2, ncols=2, figsize=(20, 10))
    #display user teams performance at home
    sns.countplot(ax=ax[0][0],x='HomeTeam', hue='FTR', data=data1, palette='Dark2',hue_order=['H','A','D'])
    #display user teams performance at away games
    sns.countplot(ax=ax[0][1],x='AwayTeam', hue='FTR', data=data2, palette='Dark2',hue_order=['H','A','D'])
    #display user teams performance in the recent ten games at home
    sns.lineplot(ax=ax[1][0],x='Date', y='FTR', data=data1[-10:-1], color='Slateblue', alpha=0.8, linewidth=3, ci=None)
    #display user teams performance in last ten games away
    sns.lineplot(ax=ax[1][1],x='Date', y='FTR', data=data2[-10:-1], color='Slateblue', alpha=0.8, linewidth=3, ci=None)
    # Display the graph
    plt.show()
```

```python
#Function to compare performance of two user entered teams
def teamcomparision(hometeam=None,awayteam=None):
    data1=data[(data['HomeTeam']== hometeam) & (data['AwayTeam']== awayteam)]
    plotPieChart(data=data1, column='FTR')
    plt.show()
```
